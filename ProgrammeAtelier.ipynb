```python
import cv2
import numpy as np

# Charger l'image
img = cv2.imread("C:/Users/arthu/OneDrive/Bureau/123.jpg", cv2.IMREAD_GRAYSCALE)

# Convertir l'image en niveaux de gris
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Appliquer un filtre gaussien pour réduire le bruit
blur = cv2.GaussianBlur(gray, (5, 5), 0)

# Détecter les contours en utilisant la méthode Canny
canny = cv2.Canny(blur, 50, 150)

# Détecter les cercles en utilisant la transformée de Hough
circles = cv2.HoughCircles(canny, cv2.HOUGH_GRADIENT, 1, 20,
                           param1=50, param2=30, minRadius=0, maxRadius=0)

# Dessiner les cercles détectés dans l'image originale
if circles is not None:
    circles = np.round(circles[0, :]).astype("int")
    for (x, y, r) in circles:
        cv2.circle(img, (x, y), r, (0, 255, 0), 2)

# Afficher l'image résultante
cv2.imshow("Cercles détectés", img)
cv2.waitKey(0)
```
